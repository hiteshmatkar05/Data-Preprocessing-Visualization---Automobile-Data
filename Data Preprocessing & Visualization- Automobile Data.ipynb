```python
# df['price']=df['price'].astype(str)
# df['price']=df['price'].str.replace("?",'').replace('',np.nan)
# df['price']=df['price'].astype(float)

df['price'] = df['price'].astype(str)
df['price'] = df['price'].str.replace('?', '', regex=False)
df['price'] = df['price'].replace('', np.nan).astype(float)

# Fill missing values
df['price'] = df['price'].fillna(df['price'].median())

# Convert to int
df['price'] = df['price'].astype(int)
```

# A] Vehicle Pricing Trends :- 

# Q.1) What are the major factors influencing vehicle price?

```python
# Select numerical columns
num_cols = df_new.select_dtypes(include=['int64', 'float64']).columns

# Correlation matrix
corr_matrix = df_new[num_cols].corr()

# Plot heatmap
plt.figure(figsize=(9,6))
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)

plt.title('Correlation Heatmap: Numerical Factors Influencing Vehicle Price')
plt.show()
```

```python
# Analyze the impact of horsepower on vehicle price :- 

plt.figure(figsize=(6,4))
sns.scatterplot(data = df_new, x='width', y='price')
plt.title('width vs Price')
plt.xlabel('width')
plt.ylabel('Price')
plt.show()
```

```python
# Analyze how engine size influences vehicle pricing :- 

plt.figure(figsize=(6,4))
sns.scatterplot(data=df_new, x='engine-size', y='price')
plt.title('Engine Size vs Vehicle Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.show()
```

```python
# Analyze the relationship between vehicle weight and price :- 

plt.figure(figsize=(6,4))
sns.scatterplot(data=df_new, x='curb-weight', y='price')
plt.title('Curb Weight vs Vehicle Price')
plt.xlabel('Curb Weight')
plt.ylabel('Price')
plt.show()
```

# Q.2)How does brand reputation (e.g., BMW vs. Toyota) correlate with pricing?

```python
# Calculate average price for each brand
brand_price = df_new.groupby('make')['price'].mean().sort_values(ascending=False)

brand_price
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=brand_price.index, y=brand_price.values)

plt.xticks(rotation=90)
plt.title('Average Vehicle Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.show()
```

```python
# Filter BMW and Toyota data
bmw_toyota = df_new[df_new['make'].isin(['bmw', 'toyota'])]

plt.figure(figsize=(6,4))
sns.boxplot(data=bmw_toyota, x='make', y='price')
plt.title('Price Comparison: BMW vs Toyota')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
```
